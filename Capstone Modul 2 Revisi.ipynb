Untuk lebih jelas,akan dilakukan klasifikasi berdasarkan kategori umur yang akan dibagi berdasarkan kategori usia sebagai berikut:
* 0-14 years (children/Anak-Anak)
* 15-24 years (early working age/Usia Pekerja Awal)
* 25-54 years (prime working age/ Usia Pekerja Utama)
* 55-64 years (mature working age/Usia Pekerja Dewasa)
* 65 years and over (elderly/Usia Lanjut)

```python
# Fungsi Tabel Kategori terhadap Kepuasan Penumpang
def TabelKepuasan(data,kepuasan,df):
    # Membuat crosstab
    Jml = pd.crosstab(df[data],df[kepuasan])
    Prop =pd.crosstab(df[data],df[kepuasan],normalize='index').round(4)*100
    # Menggabungkan Tabel
    Tabel_Kepuasan = pd.merge(Jml, Prop,on=data)
    Tabel_Kepuasan.insert(2,'Total',(pd.crosstab(df[data],df[kepuasan]).sort_index().sum(axis=1)))
    Tabel_Kepuasan.rename(columns={'Neutral_or_Dissatisfied.x':'Neutral_Dissatisfied.Val','satisfied.x':'Satisfied.Val', 'Neutral_or_Dissatisfied.y':'Neutral_Dissatisfied_%','Satisfied_y':'Satisfied_%'},inplace=True)
    return Tabel_Kepuasan

# Pembuatan Tabel dengan tambahan kategori usia
def KategoriUsia (Umur): 
    if Umur < 15: 
        return 'Children'
    elif (Umur >= 15) & (Umur < 25): 
        return 'Early Working Age'
    elif (Umur >= 25) & (Umur < 55):
        return 'Prime Working Age'
    elif (Umur >= 55) & (Umur < 65):
        return 'Mature Working Age'
    elif Umur>=65:
        return 'Elderly'
```

```python
# Tabel proporsi umur
usia = df3.copy()
    
usia['KategoriUsia'] = usia['age'].apply(lambda Umur: KategoriUsia(Umur))

# Proporsi Umur terhadap Kepuasan Penumpang
Proporsi = TabelKepuasan('KategoriUsia','satisfaction',usia)
Proporsi
```

```python
# Visualisasi proporsi berdasarkan umur secara keseluruhan
fig=px.sunburst(usia,path=['KategoriUsia','satisfaction'],color='satisfaction', title=f' Prorporsi Usia Keseluruhan Berdasarkan Usia')
fig.show()

# Visualisasi proporsi berdasarkan umur per kategori
plt.style.use('seaborn')
ct=pd.crosstab(usia['KategoriUsia'],usia['satisfaction'],normalize='index').sort_index()
ct.plot(kind='bar', stacked=True)
plt.title('Prorporsi Usia Keseluruhan Berdasarkan Usia', fontsize=16)
plt.ylabel('Prop')
```

Kesimpulan yang dapat diambil dari analisis di atas adalah:
    
    
* Segmen penumpang berdasarkan umur paling banyak adalah kelompok Prime Working Age, atau kelompok Usia pekerja utama.
